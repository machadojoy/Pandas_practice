```python
def last_four_digits(num):
    return str(num)[-4:]
```

```python
df["last_four_numbers"] = df["CC Number"].apply(last_four_digits)
```

```python
df.head()
```

```python
df["total_bill"].mean()
```

```python
def determine_price(price):
    if price < 10:
        return "$"
    elif price >= 10 and price < 30:
        return "$$"
    else:
        return "$$$"
```

```python
df["price_range"] = df["total_bill"].apply(determine_price)
```

```python
df.head()
```

```python
df["total_bill"].apply(lambda num: num*2)
```

```python
def tip_quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "generous"
    else:
        return "other"
tip_quality(16.99, 1.01)
```

```python
df["tip_quality"] = df[["total_bill", "tip"]].apply(lambda df: tip_quality(df["total_bill"], df["tip"]), axis=1) # axis=1 is the column
```

```python
df[["tip_quality", "total_bill", "tip"]]
```

```python
#np.vectorize() works faster and is easier to use in place of apply when more than 1 column is used
df["tip_quality"] = np.vectorize(tip_quality)(df["total_bill"], df["tip"]) 
```
