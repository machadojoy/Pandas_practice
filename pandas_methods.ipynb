```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("tips.csv")
```

```python
def last_four_digits(num):
    return str(num)[-4:]
```

```python
df["last_four_numbers"] = df["CC Number"].apply(last_four_digits)
```

```python
df.head()
```

```python
df["total_bill"].mean()
```

```python
def determine_price(price):
    if price < 10:
        return "$"
    elif price >= 10 and price < 30:
        return "$$"
    else:
        return "$$$"
```

```python
df["price_range"] = df["total_bill"].apply(determine_price)
```
